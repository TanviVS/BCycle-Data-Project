```python
#UR1
#Importing all the necessary libraries for our files
import csv as csv
%matplotlib inline
import matplotlib.pyplot as plt

#Using an in built python function to open reader 
with open('Austin BCycle Trips.csv', 'r') as csvfile:
    #Initializing our reader
    reader = csv.DictReader(csvfile)
    
    #Initializing our variables 
    monday_count = 0
    tuesday_count = 0
    wednesday_count = 0
    thursday_count = 0
    friday_count = 0
    saturday_count = 0
    sunday_count = 0
    
    #Reading each specific row in a specified column
    for row in reader:
        #Specifying column we are going to read
        day_of_week = row['day of week']
        
        #We are increasing counts based on the day of the week the reader reads within the excel cell
        if (day_of_week == "Monday"):
            monday_count +=1
    
        elif (day_of_week == "Tuesday"):
            tuesday_count +=1
    
        elif (day_of_week == "Wednesday"):
            wednesday_count +=1
    
        elif (day_of_week == "Thursday"):
            thursday_count +=1
    
        elif (day_of_week == "Friday"):
            friday_count +=1
    
        elif (day_of_week == "Saturday"):
            saturday_count +=1
    
        elif (day_of_week == "Sunday"):
            sunday_count +=1
            
    #Storing all count variables into a list to measure frequency of trips by days of the week
    num_trips_day_of_week = [monday_count, tuesday_count, wednesday_count, thursday_count, friday_count, saturday_count, sunday_count]
    #Our x axis is simply labels of the days of the week
    left_edges = ["Mon", "Tues", "Wed", "Thurs", "Fri", "Sat", "Sun"]
    
    #Set bar width so everything can fit in 
    bar_width = .75
    
    #Add title to the graph
    plt.title('Number of BCycle trips for each day of the week')

    # Add labels to the axes.
    plt.xlabel('Day of the week')
    plt.ylabel('Number of BCycle trips')
    
    #Brief description
    print("Explanation: Since there are a defined number of days of the week that are globally well- known, instead of ")
    print("             creating a dictionary for this problem, I decided to increment counts for each day of the week, ")
    print("             and then graphed their lists accordingly.")
    
    #Graph the bar graph and define a color and width
    plt.bar(left_edges, num_trips_day_of_week, bar_width, color = ('xkcd:azure'))
    #Display the graph
    plt.show()

```

```python
#UR3
#Importing all the necessary libraries for our files
import operator
import csv as csv
%matplotlib inline
import matplotlib.pyplot as plt

#Numpy is a scientific mathematics library that will allow us to create arrays for this problem 
import numpy as np

#Both of these libraries will be used for predicting the models
#Train test split will train our values while polynomial features will specifically be used to project the line 
#The benefit of using polynomial features is that we can easily change the number of maximum degrees we want our line to have
#Mean squared error and r2score will give us a quantity on how much this line matches the points telling us if it is a good fit
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

#Using an in built python function to open reader 
with open('Austin BCycle Trips.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    #Initializing our count variables
    count_2014 = 0 
    count_2015 = 0
    count_2016 = 0
    count_2017 = 0
    
    #We are increasing counts based on the year the reader reads within the excel cell
    #We remove 2017 data since we are still predicting the data
    for row in reader:
        year = int(row['year'])
        
        if (year == 2014):
            count_2014 +=1
            
        elif (year == 2015):
            count_2015 +=1
                    
        elif (year == 2016):
            count_2016 +=1

#Define our x and y arrays to train our model
#The x array is using the reshape function to become a 2D array as that is a requirement for the model
x = np.array([1, 2, 3]).reshape((-1, 1))
y = np.array([count_2014, count_2015, count_2016])

#We set the polynomial degree to 1 so the graph will be linear
polynomial_features= PolynomialFeatures(degree=1)
#Trains x values into polynomial features
x_poly = polynomial_features.fit_transform(x)

#Defines model to fit under a linear regression model
model = LinearRegression()
#Trains model under given points y points
model.fit(x_poly, y)
#Defines new y predicition values based on the trained model and trained x values
y_poly_pred = model.predict(x_poly)

#This two lines of code will quantify how close this line is to a line of best fit
rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)

#Assumes array as input
sort_axis = operator.itemgetter(0)
#Builds function
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
#Redefines predicted models based on sorted training model
x, y_poly_pred = zip(*sorted_zip)

#Initializaing variables to calculate actual line
y1 = y_poly_pred[0]
y2 = y_poly_pred[1]
x1 = x[0]
x2 = x[1]

#Slope equation
m = (y2 - y1)/(x2 - x1)
#Y- intercept equation
b = y1 - (m * x1)

#Initializing our fourth variable for 2017 graph
x4 = 4
#Equation for 2017 prediction using calculations above
count_2017 = (m* x4) + b

#We have to redefine our variables since we cannot edit our arrays
x = [1, 2, 3, x4]
y = [count_2014, count_2015, count_2016, count_2017]

#Graph title
plt.title('Number of BCycle trips Projected in 2017')

# Add labels to the axes.
plt.xlabel('Year')
plt.ylabel('Number of BCycle trips')

#X axis labels
plt.xticks([1, 2, 3, 4],
               ['2014', '2015', '2016', '2017'])

#Brief Explanation
print("Explanation:For this problem, I used a combination of two libraries: A Linear Regression library and a Polynomial")
print("            Shapes library. I then used the data that I found in UR2 (besides the 2017 data we had) to train ")
print("            data points for our newly defined line of best fit. I used these predicitve points to calculate")
print("            the slope and y- intercept for my equation and entered a x4 value to create predictions for")
print("            the amount of BCycle trips in 2017.")

#Displays graph
plt.bar(x, y, color = ('xkcd:azure'))
plt.show()
```

```python
#UR4
#IMporting necessary libraries for files
import csv as csv
%matplotlib inline
import matplotlib.pyplot as plt

#Initializing all dictionaries, lists, and variables needed
bcycle_trips_list = []
subscription_name_list = []
subscriber_type_dict = {}
i = 0

#Using an in built python function to open reader 
with open('Austin BCycle Trips.csv', 'r') as csvfile:
        #Initializing our reader
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        #Reading specific column data that we need for this specific problem
        subscriber_type = row['subscriber type']    
        
        #If the subscriber type has already been recorded once in the dictionary, we are just going to increment the count
        if subscriber_type in subscriber_type_dict:
            count = subscriber_type_dict[subscriber_type]
            count += 1
            
        #Initialize the count if the subscriber type is being recorded for the first time
        else:
            count =1
            
        #Recording both the key and value inside the dictionary
        subscriber_type_dict[subscriber_type] = count

#Defines how many subscription types we want to actually graph
num_subscriber_type = int(input("Enter the number of types of subscriptions you would like to graph: "))


#Runs loop until user- specified variable above is reached
while (i != num_subscriber_type):
    #Essentially makes sure we do not leave out or record extra variables
    #Makes sure we do not have an infinite loop
    subscriber_type_name = input("Please enter the name of the subscription: ")
    
    if subscriber_type_name in subscriber_type_dict:
        #Error Handling
        if subscriber_type_name in subscription_name_list:
            print("You have already entered this subscription type. Please try again")
            
        #Records subscription type and count in seperate lists and increments the loop
        else:
            count = subscriber_type_dict[subscriber_type_name]
            subscription_name_list.append(subscriber_type_name)
            bcycle_trips_list.append(count)
            i += 1
    #Error handling
    else:
        print("This subscription does not exist. Please try again")
        
#Graph title
plt.title('Number of BCycle trips per subscriber type')

# Add labels to the axes.
plt.xlabel('Subscription Type')
plt.ylabel('Number of BCycle trips')
#Changes bar graph size
fig = plt.gcf()
fig.set_size_inches(12, 8)

#Brief Explanation
print()
print("Explanation: Because I did not know all the different BCycle subscriptions that existed, I decided to create a ")
print("             dictionary that would record all variable types within the dataset. I then allowed users to pick ")
print("             which subscription type they were looking for so they would not be overloaded with data.")


#Displays graph
plt.bar(subscription_name_list, bcycle_trips_list, color = ('xkcd:azure'))
plt.show()
```

```python
#UR8
#Importing all the necessary libraries needed for these files
import csv as csv
import matplotlib.pyplot as plt

#Initializing dictionaries
area_dict = {}

#Using an in built python function to open reader 
with open('Austin BCycle Trips.csv', 'r') as csvfile:
    #Initializing the reader
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        #Reading dataset from specified column
        start_area = row['start station name']
        
        #If the start area has already been recorded once in the dictionary, we are just going to increment the count
        if start_area in area_dict:
            count = area_dict[start_area]
            count+=1
            
        #Initialize the count if the start area is being recorded for the first time
        else:
            count =1
                    
        #Recording both the key and value inside the dictionary
        area_dict[start_area] = count
        
with open('Austin BCycle Trips.csv', 'r') as csvfile:
    #Initializing the reader
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        #Reading dataset from specified column
        end_area = row['end station name']
        
        #If the end area has already been recorded once in the dictionary above, we are just going to increment the count
        #We are simply going to be adding in these areas to the same dictionaries so we can get the total count
        if end_area in area_dict:
            count = area_dict[end_area]
            count+=1
     
        #Initialize the count if the area is being recorded for the first time
        else:
            count =1
            
        #Recording both the key and value inside the dictionary
        area_dict[end_area] = count
        
#Brief Explanation
print("Explanation: Since there was a wide range of start and end stations, I decided to record and increment all the")
print("             occurences within the same dictionary. However, a dictionary needs a special function to find a ")
print("             maximum value, which is why I used the lambda key and max function.")
print()

#Finding the maximum frequency within the dictionary using the count as the key
max_count = max(area_dict, key=lambda count: area_dict[count])
print("The most popular station is", max_count)
```
